```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#regresionLineal
```

```python
x = np.array([-1,0,1,2])
y = np.array([0,1,2,1])
plt.plot(x,y,"bo")
```

```python
from scipy.stats import linregress
```

```python
res= linregress(x,y)
res
```

```python
xn = np.arange(-1,2,0.05)
yn = res.intercept + res.slope*xn
```

```python
plt.figure(figsize=(8,6))

plt.plot(x,y,"*")
plt.plot(xn,yn)
plt.plot(x,y,"*b")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.grid()
plt.show()
```

```python
#Actividad
```

```python
t = np.array([56.5,113, 181,214.5])
p = np.array([1,5,20,40])
plt.plot(t,p,"bo")
```

```python
res= linregress(t,p)

tn = np.arange(50,250,25)
pn = res.intercept + res.slope*tn
```

```python
plt.figure(figsize=(8,6))

plt.plot(p,t,"*")
plt.plot(pn,tn)
plt.plot(p,t,"*b")
plt.xlabel("$p$")
plt.ylabel("$t$")
plt.grid()
plt.show()
```

```python
from scipy.optimize import curve_fit
```

```python
def acetona (x,a,b,c):
    return a+b*x+c*(x**2)

a, b, c =2.5,1.3,0.5
t = np.array([56.5,113, 181,214.5])
p = np.array([1,5,20,40])

(A,B,C),_= curve_fit(acetona,p,t)
print (A, B, C)

nxdat = np.arange(0,40,0.1)
ydatAjuste =acetona(nxdat, A,B,C)

plt.plot(p,t,"o",nxdat,ydatAjuste)
plt.legend(["Datos","Ajuste"],loc="best");
```
