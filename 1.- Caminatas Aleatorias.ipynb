```python
#caminar un paso hacia atras o delante dependiendo del lanzamiento de una moneda

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
np.random.seed(22)
x_0 = 0 #inicia el caminante
elementos =[1,-1] #aguila o sol
probabilidades = [0.5,0.5] #probabilidades de que se aguila o sol
delta_x = np.random.choice(elementos,10,p=probabilidades)
delta_x
```

```python
delta_x
```

```python
#generar hitorial
np.cumsum(delta_x) #Suma todos los resultados de la probabilidad
```

```python
y = np.ones(len(delta_x))
plt.plot(np.cumsum(delta_x),y)
```

```python
def walker ():
    global x,y
    eleccion =random.randint(0,3)
    if eleccion ==0:
        x+=1
    elif eleccion==1:
        x-=1
    elif eleccion==2:
        y+=1
    else:
        y-=1
    return [x,y]

x=0 #inicia el caminador
y=0
Posicion=np.zeros((2,100))
for i in np.arange(0,len(Posicion[0]),1):
    Posicion[:,i] = walker()

plt.plot(Posicion[0,:],Posicion[1,:],c="0123")
plt.plot(Posicion[0,-1],Posicion[1,-1],"ro-")
plt.plot(Posicion[0,0],Posicion[1,0],"go-")
```

```python
#Condiciones iniciales de nuestra caminata aleatoria
X_0 = np.array([[0],[0]]) #coordenada
delta_X= np.random.uniform(-1,1,(2,1000)) #
X=np.concatenate((X_0, np.cumsum(delta_X,axis=1)),axis=1)

plt.plot(X[0],X[1],c="0123")
plt.plot(X[0,-1],X[1,-1],"ro-")
plt.plot(X[0,0],X[1,0],"go-")
```
