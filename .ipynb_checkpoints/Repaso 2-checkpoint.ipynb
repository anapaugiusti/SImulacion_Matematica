$$\frac{dy}{dx} = -5y $$ 

```python
def f(y,t):
    return -5*y

t = np.linspace(0,2,50) #de 0 a 2 y seran 50 valores

for y0 in np.arange(0,10,2):
    y = odeint(f,y0,t)
    plt.plot(t,y);
```

$$\frac{dy}{dx} = y $$ 

$$\frac{dy}{dx} = -x-0.2y $$ 

```python
def f(s,t):
    x,y = s
    xdot = y
    ydot = -x -(0.2*y)
    return xdot,ydot

t = np.linspace(0,50,200)
s0 = np.array([0.25,0.25])

s = odeint(f,s0,t)
xs, ys = s.T
plt.plot(xs,ys)
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.figure(figsize=(8,8))

x = np.linspace(-1,1,10)
y = np.linspace(-1,1,10)
xx,yy =np.meshgrid(x,y)

xxdot,yydot = f([xx,yy],0)
plt.quiver(xx,yy,xxdot,yydot)
```

```python
plt.figure(figsize=(8,8))
x = np.linspace(-1,1,10)
y = np.linspace(-1,1,10)
xx,yy =np.meshgrid(x,y)
xxdot,yydot = f([xx,yy],0)
plt.quiver(xx,yy,xxdot,yydot)

t = np.linspace(0,50,200)
#s0 = np.array([0.25,0.25])
s0 = np.array([1,1])
s = odeint(f,s0,t)
xs, ys = s.T
plt.plot(xs,ys)
```

```python
plt.figure(figsize=(8,8))
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.streamplot(xx,yy,xxdot,yydot)

s = odeint(f,s0,t)
xs, ys = s.T
plt.plot(xs,ys,"b")
```

```python
plt.figure(figsize=(10,8))
plt.xlim(-1,1)
plt.ylim(-1,1)
mag= np.sqrt(xxdot**2+yydot**2)
plt.streamplot(xx,yy,xxdot,yydot,color=mag,cmap="autumn")
plt.colorbar()
```

$$\frac{dy}{dx} = x + y$$
$$\frac{dy}{dx} = x - y $$

```python
def f2(s,t):
    x,y = s
    xdot = x + y
    ydot = x - y
    return xdot,ydot

t = np.linspace(0,50,200)
s0 = np.array([0.5,-1])

s = odeint(f2,s0,t)
xs, ys = s.T
plt.plot(xs,ys)
plt.xlabel("x")
plt.ylabel("y")
```
