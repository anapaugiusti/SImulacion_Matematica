1.- JUEGO DE AZAR

Un amigo le propone el siguiente juego:
 - Usted lanzar√° dos dados (sin alterar)
 - Si el resultado
 - Su amigo le pagara 100 multiplicados por el producto, si caen dos "Unos"
 - En caso contrario te paga 200
 

```python
Dado1 = np.arange(1,7)
Dado2 = np.arange(1,7)

df = pd.DataFrame(index=np.arange(1,7), columns=["Dado1", "Dado2","Probabilidad","Ganancia"], dtype=float)

df.index.name = "Evento"
df["Dado1"] = Dado1
df["Dado2"] = Dado2
df["Ganancia"] = Dado1*Dado2*100
df["Probabilidad"] = 1/36
df
```

```python
#PROBABILIDAD DE GANAR
df.Probabilidad.sum()
```

```python
#PROBABILIDAD DE PERDER
1-df.Probabilidad.sum()
```

```python
#juego repetitivo
sum(df["Probabilidad"]*df["Ganancia"])+(1-df.Probabilidad.sum())*-200
```

```python
def Juego (Dado1,Dado2):
    if Dado1==Dado2:
        return 100*(Dado1*Dado2)
    else:
        return -200
Dado1 = np.random.randint(1,7)
Dado2 = np.random.randint(1,7)
Juego(Dado1,Dado2)
```

```python
N = 200
SimularJuego1 = np.zeros(N)

for i,j in enumerate (SimularJuego1):
    Dado1 = np.random.randint(1,7)
    Dado2 = np.random.randint(1,7)
    SimularJuego1[i] = Juego(Dado1,Dado2)

plt.hist(SimularJuego1,bins=10);
print(np.mean(SimularJuego1))
```

```python
plt.plot(np.cumsum(SimularJuego1))
#SUMA DE GANANCIAS Y PERDIDAS
```

MEJORAR CODIGO

```python
N= 10000
win=100
lose=-340

Dados=np.random.randint(1,7,(N,2))
jugador1 = np.where(Dados[:,0] == Dados[:,1],Dados[:,0]*Dados[:,1]*win,lose)
plt.plot(np.cumsum(jugador1))
```

```python
for i in range (20):
    Dados=np.random.randint(1,7,(N,2))
    jugador1 = np.where(Dados[:,0] == Dados[:,1],Dados[:,0]*Dados[:,1]*win,lose)
    plt.plot(np.cumsum(jugador1))
#gananado cliente po N mayor
#menos lose, cliente pierde mas
```
